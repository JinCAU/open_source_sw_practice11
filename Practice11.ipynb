### 1. importing required libraries and loading the iris plants dataset (classification)
- libraries: datasets, model_selection, metrics, tree, ensemble

```python
import sklearn
import sklearn.datasets
import sklearn.model_selection
import sklearn.metrics
import sklearn.tree
import sklearn.ensemble

iris = sklearn.datasets.load_iris()
X, y = sklearn.datasets.load_iris( return_X_y=True )
```

### 2. dividing the datasets into two parts i.e. training datasets and test datasets

```python
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.3)
```

### 3. creating a random forest classifier 

```python
rfc=sklearn.ensemble.RandomForestClassifier()
```

### 4. training the model on the training dataset 

```python
rfc.fit(X_train, y_train)
y_pred=rfc.predict(X_test) 
```

### 5. performing predictions on the test dataset and measuring the performance
- metrics are used to find accuracy or error

```python
print("Accuracy : %.2f" %sklearn.metrics.accuracy_score(y_test, y_pred))
```

### 6. visualizing all trees
- RandomForestClassifier.estimators_
- tree.plot_tree

```python
import matplotlib.pyplot as plt

# plt.figure(figsize=(20,20))
# rfc.estimators_ = sklearn.tree.plot_tree(rfc.estimators_[0], filled=True)

fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(20, 4))
for index in range(0, 5):
    sklearn.tree.plot_tree(rfc.estimators_[index], ax=axes[index])
plt.show()
```

### 7. creating make moons dataset, splitting them, and training the model on the training dataset
- datasets.make_moons

```python
#X, y=sklearn.datasets.make_moons( return_X_y=True )
# make_moons=sklearn.datasets.make_moons()
# X, y = make_moons(100, noise=0.035, random_state=20)
# print(X)
#print(y.shape)
#X_train, X_test, y_train, y_test=sklearn.model_selection.train_test_split(X, y, test_size=0.3)

X, y= sklearn.datasets.make_moons(noise=0.25)
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.3)
rfc=sklearn.ensemble.RandomForestClassifier()
rfc.fit(X_train, y_train)
y_pred=rfc.predict(X_test)
print("Accuracy : %.2f" %sklearn.metrics.accuracy_score(y_test, y_pred))
```

### 8. visualizing 4 individual trees from the trained random forest
- from sklearn.inspection import DecisionBoundaryDisplay

```python
from sklearn.inspection import DecisionBounderyDisplay

fig, axes = plt.subplots(nrows = 1,ncols = 5,figsize = (20,4))
for index in range(0, 4):
    DecisionBoundaryDisplay.from_estimator(
            rfc.estimators_[index], X, alpha=0.5, response_method="predict", ax=axes[index]
        )
    axes[index].scatter(X[:,0],X[:,1],c=y)
    axes[index].set_title('Tree: ' + str(index))

DecisionBoundaryDisplay.from_estimator(
        rfc, X, alpha=0.5, response_method="predict", ax=axes[-1]
    )
axes[-1].scatter(X[:,0],X[:,1],c=y)
axes[-1].set_title('Forest: ' + str(index))    
plt.show()
```

### 9. 
1. load wine dataset from scikit learn, 
2. train both single decision tree and a random forest
3. compare their accuracy

```python
X, y= sklearn.datasets.load_wine( return_X_y=True)
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.3)
rfc=sklearn.ensemble.RandomForestClassifier()
rfc.fit(X_train, y_train)
y_pred=rfc.predict(X_test)
print("Accuracy : %.2f" %sklearn.metrics.accuracy_score(y_test, y_pred))

fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(20, 4))
for index in range(0, 5):
    sklearn.tree.plot_tree(rfc.estimators_[index], ax=axes[index])
plt.show()
```

### 11. 
1. load Olivetti faces dataset from scikit learn
    - sklearn.datasets.fetch_olivetti_faces
2. train two random forests with different hyperparameters
3. visualize them
4. compare their accuracy

```python
X, y= sklearn.datasets.fetch_olivetti_faces(return_X_y=True)
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.3)
rfc=sklearn.ensemble.RandomForestClassifier(max_features="log2")
rfc.fit(X_train, y_train)
y_pred=rfc.predict(X_test)
print("Accuracy : %.2f" %sklearn.metrics.accuracy_score(y_test, y_pred))

fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(20, 4))
for index in range(0, 5):
    sklearn.tree.plot_tree(rfc.estimators_[index], ax=axes[index])
plt.show()
```
